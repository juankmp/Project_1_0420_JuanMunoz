```python
import pandas as pd
```

```python
#Importing the CSV processed v1
path_from= '/home/usuario/Documentos/Ironhack/Project_1_0420_JuanMunoz/data/processed'


df_all_raw_data = pd.read_csv(f'{path_from}/df_all_raw_data.csv', sep=';') 
data_jobs_api = pd.read_csv(f'{path_from}/data_jobs_api.csv', sep=';') 
data_country_scraping = pd.read_csv(f'{path_from}/data_country_scraping.csv', sep=';') 


df_all_raw_data.head()
```

```python
df_1 = pd.merge(df_all_raw_data,data_jobs_api, how='left',on = 'normalized_job_code')
df_all = pd.merge(df_1,data_country_scraping, how='left',on = 'country_code')

display(df_all[0:5])
```

```python
#Changing column names

new_names_list = ['uuid','education_level','full_time_job','job_code',
'uuid_1','quest_awareness',
'quest_vote',
'quest_effect',
'quest_arguments_for',
'quest_arguments_against',
'uuid_2',
'country_code',
'rural',
'uuid_3',
'age',
'gender',
'has_children',
'age_group',
'job_title',
'job_title_normalized',
'country',
'population']

df_all.columns = new_names_list
```

```python
#filtering data needed

filter_list = ['uuid','education_level','full_time_job','job_code',
'job_title',
'country_code',
'country',
'rural',
'age',
'gender',
'has_children',
'age_group',
'quest_awareness',
'quest_vote',
'quest_effect',
'quest_arguments_for',
'quest_arguments_against']

df_all = df_all[filter_list]
```

```python
df_all
```

```python
#Changing in education_level 

df_all['education_level'] = df_all['education_level'].fillna('undefined')
df_all['job_code'] = df_all['job_code'].fillna('Unemployed')
df_all['job_title'] = df_all['job_title'].fillna('Unemployed')

#function for remove strange characters
def replace(x):
    try:
        return x.replace("‰û_ ", "").replace("‰Û_ ", "")
    except:
        return x 
    
def lower(x):
    try:
        return x.lower()
    except:
        return x 
```

```python
df_all['quest_effect']= df_all['quest_effect'].apply(replace)
```

```python
df_all['rural'] = df_all['rural'].apply(lower)
df_all['rural'].value_counts()
```

```python
def replace_age(x):
    try:
        return int(x.replace(' years old',""))
    except:
        return x
    
df_all['age'] = df_all['age'].apply(replace_age) 


def calculate_age(x):
    try:
        if len(str(x))>3:
            reference_age = 2016
            real_age = reference_age - x
            return real_age
        else:
            return x
    except:
        return x

df_all['age'] = df_all['age'].apply(calculate_age)

 #because from 1991 is in the age_group 25 and 1990 is in age 26 - CALCULATE THE AGE
```

```python
df_all['age'].value_counts()
```

```python
df_all['has_children'] = df_all['has_children'].apply(lower)
df_all['has_children'].value_counts()
```

```python
df_all['age_group'] = df_all['age_group'].replace("juvenile",'14_25')
df_all['age_group'].value_counts()
```

```python
df_all['gender'] = df_all['gender'].apply(lower).replace('fem','female')
df_all['gender'].value_counts()
```
